```python
# Data Load and Check
import pandas as pd

df = pd.read_csv("Nassau Candy Distributor.csv")

print(df.shape)
df.head()
```

```python
# Data Cleaning here

# Remove invalid data
df = df[df['Sales'] > 0]

# Fill missing values
df['Units'] = df['Units'].fillna(1)

# Check nulls
print(df.isnull().sum())
```

```python
df['Gross Margin %'] = (df['Gross Profit'] / df['Sales']) * 100
df['Profit per Unit'] = df['Gross Profit'] / df['Units']

df.head()
```

```python
# Top profit Product show here

top_products = df.groupby('Product Name')['Gross Profit'].sum().sort_values(ascending=False)
top_products.head(10)
```

```python
# YaHa Low Margin Products show here

low_margin = df[df['Gross Margin %'] < 10]
low_margin.head()
```

```python
# High Sales but Low Profit

danger_products = df[(df['Sales'] > df['Sales'].mean()) & (df['Gross Margin %'] < 10)]
danger_products
```

```python
# Division Analysis here...

division = df.groupby('Division').agg({
    'Sales': 'sum',
    'Gross Profit': 'sum',
    'Gross Margin %': 'mean'
})

division
```

```python
# Pareto Analysis here...

df_sorted = df.sort_values(by='Gross Profit', ascending=False)

df_sorted['Cumulative Profit'] = df_sorted['Gross Profit'].cumsum()

total_profit = df['Gross Profit'].sum()

df_sorted['Profit %'] = (df_sorted['Cumulative Profit'] / total_profit) * 100

df_sorted[['Product Name', 'Profit %']].head(20)
```

```python
# Visualization here...

import seaborn as sns
import matplotlib.pyplot as plt

# Scatter Plot here...


sns.scatterplot(data=df, x='Sales', y='Gross Profit')
plt.title("Sales vs Profit")
plt.show()
```

```python
# Top products chart here...


top_products.head(10).plot(kind='bar')
plt.title("Top 10 Products by Profit")
plt.show()
```

```python
#Its ML Part Here...

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df[['Cost', 'Units']] # x means input 
y = df['Gross Profit']    # y means output

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

print("Train Score:", model.score(X_train, y_train))
print("Test Score:", model.score(X_test, y_test))
print("Model Coefficients:", model.coef_)
```
